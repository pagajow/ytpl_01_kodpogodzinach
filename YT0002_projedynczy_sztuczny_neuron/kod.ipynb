```python
from neuron import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
dane, wyniki  = daneTestowe(100)
n = Neuron(liczbaWejsc=len(dane[0]), funkcjaAktywacji=None)
h = n.trenuj(daneWejsciowe=dane, daneWynikowe=wyniki, epoki=1000, tempoUczenia=0.001)
```

```python
bledy = np.array([100*x[2]/x[0] for x in h[-1] if x[0] != 0])

plt.bar(range(len(bledy)), bledy)
plt.title("Błędy względne predykcji")
plt.xlabel('Nr')
plt.ylabel('Błąd (%)')
```
